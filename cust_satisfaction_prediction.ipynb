```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(x='satisfaction', data=df)

plt.title('Distribution of Customer Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('Number of Passengers')
```

```python
sns.countplot(x='Gender',hue='satisfaction',data=df)
plt.title('Gender vs Satisfaction')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.legend(title='Satisfaction')
plt.show()
```

```python
sns.countplot(x='Customer Type', hue='satisfaction', data=df)
plt.title('Customer Type vs Satisfaction')
plt.xlabel('Customer Type')
plt.ylabel('Number of Passengers')
```

```python
sns.countplot(x='Type of Travel', hue='satisfaction', data=df)
plt.title('Type of Travel vs Satisfaction')
plt.xlabel('Type of Travel')
plt.ylabel('Number of Passengers')
```

```python
sns.countplot(x='Class', hue='satisfaction', data=df)

plt.title('Passenger Class vs Satisfaction')
plt.xlabel('Travel Class')
plt.ylabel('Number of Passengers')
```

```python
sns.boxplot(x='satisfaction', y='Age', data=df)

plt.title('Age by Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('Age')
```

```python
avg_dist = df.groupby('satisfaction')['Flight Distance'].mean()
avg_dist.plot(kind='bar')
plt.title('Average Flight Distance by Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('Average Flight Distance')
```

```python
sns.boxplot(x='satisfaction',y='Inflight entertainment',data=df)
plt.title('Inflight Entertainment Rating by Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('Entertainment Rating')
```

```python
sns.boxplot(x='satisfaction',y='Baggage handling',data=df)
plt.title('Baggage Handling Rating by Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('Baggage Handling Rating')
```

```python
sns.boxplot(x='satisfaction',y='Cleanliness',data=df)

plt.title('Cleanliness Rating by Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('Cleanliness Rating')
```

```python
sns.barplot(x='satisfaction',y='Departure Delay in Minutes',data=df)

plt.title('Departure Delay by Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('Departure Delay (Minutes)')
```

```python
sns.barplot(x='satisfaction',y='Arrival Delay in Minutes',data=df)

plt.title('Arrival Delay Vs Satisfaction')
plt.xlabel('Satisfaction')
plt.ylabel('Arrival Delay (Minutes)')
```

```python
num_cols = [
    'Age',
    'Flight Distance',
    'Inflight entertainment',
    'Baggage handling',
    'Cleanliness',
    'Departure Delay in Minutes',
    'Arrival Delay in Minutes'
]

sns.heatmap(df[num_cols].corr(),
            annot=True,
            cmap='coolwarm')

plt.title('Correlation Heatmap')
```
